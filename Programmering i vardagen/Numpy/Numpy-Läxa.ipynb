# 3.
Skriv ett program som ritar grafen till funktionen f(x) = 3x+1 för 0 <= x <= 10. Namnge x-axeln med "x" och y-axeln med "y" och använd en lämplig titel. Tips: använd funktionen linspace() i numpymodulen och modulen matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10)

def f(x):
    return 3*x+1

plt.plot(x,f(x))
plt.title("Funktion för kx + m")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# 4.
Skriv ett program som ritar grafen till funktionen f(x) = x^2-1 för -2 <= x <= 2. Namnge x-axeln med "x" och y-axeln med "y" och använd en lämplig titel. Tips: använd funktionen linspace() i numymodulen och modulen matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2)

def f(x):
    return x**2 -1

plt.plot(x,f(x))
plt.title("Funktion")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# 5.
En raket har en rörelsebana som beskrivs av h(t) = -4t^2 + 24t + 1, h(t) är höjden i meter och t är tiden i sekunder. Rita rörelsebanan och beräkna högsta höjden. Tips: använd matplotlib och linspace i numpymodulen.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,6)

def h(t):
    return -4*t**2 + 24*t + 1

maximi = np.max(h(t))
print(f"Maximipunkt: {maximi}")

plt.plot(t,h(t))
plt.plot(3,maximi,'ro')
plt.title("Åkbana för raket")
plt.xlabel("tid (s)")
plt.ylabel("höjd (m)")
plt.show()
```
