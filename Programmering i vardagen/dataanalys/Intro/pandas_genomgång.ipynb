# Pandas grunder

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Matteprov.csv")  # df står för "dataframe"

#print(df.head(3))
#print(df.tail(4))
#print(df.dropna()) # slänger raderna som innehåller NaN
df = df.fillna(0)   # byter ut NaN mot 0:or
df["Total"] = df.sum(axis=1)
df["Betyg"] = np.where(df["Total"] > 7, "Godkänt", "Underkänt")
df
```

## Select data
- rader
- kolumner

```python
# plocka ut kolumner med kolumnnamn
df["Betyg"]
```

## Conditional Filtering
- villkorlig sökning (filtrering)

```python
df_G = df[df["Betyg"] == "Godkänt"]
df_U = df[df["Betyg"] == "Underkänt"]
df_U
```

## Datavisualisering

```python
import matplotlib.pyplot as plt

betygsLista = df["Betyg"].value_counts().tolist()
labels = df["Betyg"].unique()

plt.bar(labels, betygsLista, width=.4, color = ['purple', 'orange'])
plt.title('Antal godkända/underkända i klassen')
plt.show()
print(betygsLista)
print(labels)
```
