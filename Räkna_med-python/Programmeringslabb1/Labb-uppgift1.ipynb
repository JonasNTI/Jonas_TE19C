```python
# (d) Simulera många punkter slumpmässigt mellan -1 och 1 för x- och y-koordinater, genom att använda funktionen random.uniform(-1,1).

import random as rnd
import matplotlib.pyplot as plt
import numpy as np

plt.grid(linestyle='-')

for i in range(100):
    x = rnd.uniform(-1,1)
    y = rnd.uniform(-1,1)
    plt.plot(x,y, '.r')
plt.show()
```

```python
# (e) Räkna antalet punkter som hamnar inne i cirkeln och dividera med antalet simulerade punkter.

import matplotlib.pyplot as plt
import random as rnd
import math

total = 100000
blue = 0
red = 0
#simulerar punkter efter värdet av "total"
for i in range(total):
    x = rnd.uniform(-1,1); 
    y = rnd.uniform(-1,1); 
    #räknar ut om punkten är i eller utanför cirkeln
    r = math.sqrt(x**2 + y**2)
    if r < 1:
        blue += 1
    else:
        red += 1
    
#Dividerar simulerade punkter med totala antalet punkter.
div = blue/total

print(f"{div}")
```

```python
# (f) Multiplicera resultatet med 4 och skriv ned kommentarer i koden på varför du fick det resultatet.


Ans = div * 4
print(f"{Ans:.2f}")
```

```python
# (g) Använd matplotlib och rita punkterna som hamnar inne i cirkeln och punkterna som hamnar utanför. Använd två olika färger för att särskilja dem åt.

import matplotlib.pyplot as plt
import random as rnd
import math

fig, ax = plt.subplots(1)
ax.set_aspect(1)

for i in range(1000):
    x = rnd.uniform(-1,1); 
    y = rnd.uniform(-1,1); 
    r = math.sqrt(x**2 + y**2)
    if r < 1:
        plt.plot(x,y, '.b')
    else:
        plt.plot(x,y, '.r')
plt.show()
```
